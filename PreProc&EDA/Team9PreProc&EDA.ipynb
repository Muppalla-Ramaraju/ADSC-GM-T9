Cleaning and Joining datasets

Joining datasets

````python
import pandas as pd

# Assuming household_data, person_data, vehicle_data, and trip_data are DataFrames

# Check for duplicate columns in each dataset
print("Duplicate columns in household_data:", household_data.columns.duplicated().sum())
print("Duplicate columns in person_data:", person_data.columns.duplicated().sum())
print("Duplicate columns in vehicle_data:", vehicle_data.columns.duplicated().sum())
print("Duplicate columns in trip_data:", trip_data.columns.duplicated().sum())

# Merge the datasets
df = pd.merge(household_data, person_data, on='HOUSEID', how='inner')
print("Duplicate columns:", df.duplicated().sum())
# Identify duplicate columns
duplicates = [col for col in df.columns if col.endswith(('_x'))]

# Drop the duplicate columns
df = df.drop(columns=duplicates)

# Rename columns with '_y' suffix by removing the suffix
df = df.rename(columns={col: col.replace('_y', '') for col in df.columns})


# Display the new dataframe
df

df = pd.merge(df, vehicle_data, on='HOUSEID', how='inner')
# Drop duplicate columns
df = df.loc[:, ~df.columns.duplicated()]

# Identify duplicate columns
duplicates = [col for col in df.columns if col.endswith(('_x'))]

# Drop the duplicate columns
df = df.drop(columns=duplicates)

# Rename columns with '_y' suffix by removing the suffix
df = df.rename(columns={col: col.replace('_y', '') for col in df.columns})


# Display the new dataframe
df

print("Duplicate columns:", df.duplicated().sum())
df = pd.merge(df, trip_data, on='HOUSEID', how='inner')

# Drop duplicate columns
df = df.loc[:, ~df.columns.duplicated()]

###########Final drop after all sets
# Identify duplicate columns
duplicates = [col for col in df.columns if col.endswith(('_x'))]

# Drop the duplicate columns
df = df.drop(columns=duplicates)

# Rename columns with '_y' suffix by removing the suffix
df = df.rename(columns={col: col.replace('_y', '') for col in df.columns})


# Display the new dataframe
df

# Now you have merged all DataFrames into one based on the 'HOUSEID' column and removed duplicate columns
print("Final dimensions of the merged DataFrame:", df.shape)
````
